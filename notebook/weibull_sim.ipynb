```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
def set_up(ax):
    for spine in ax.spines.values():
        spine.set_linewidth(1.2)       # เพิ่มความหนาเส้นกรอบ
        spine.set_color('black')       # สีดำคม
        spine.set_visible(True)        # ให้แน่ใจว่ากรอบทุกด้านแสดง

    # --- ปรับ tick ให้อ่านง่าย ---
    ax.tick_params(direction='in', length=4, width=1, colors='black', labelsize=16)
    ax.grid(
        visible=True,
        which='major',
        color='gray',
        linestyle='--',
        linewidth=0.6,
        alpha=0.7
    )
    ax.minorticks_on()  # เปิด tick ย่อย
    ax.grid(
        visible=True,
        which='minor',
        color='lightgray',
        linestyle=':',
        linewidth=0.5,
        alpha=0.6
    )
```

```python
df_1 = pd.read_excel('../dataset/Weibull_Data.xlsx',sheet_name='The comparison of fit theoretic')
df_2 = pd.read_excel('../dataset/Weibull_Data.xlsx',sheet_name='four parameter estimation')
```

```python
mpl.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "font.size": 12,
    "axes.labelsize": 14,
    "axes.titlesize": 14,
    "legend.fontsize": 11,
    "figure.dpi": 300,
    "axes.grid": True,
    "grid.linestyle": "--",
    "grid.alpha": 0.5,
    "lines.linewidth": 1.5,
})
fig, ax = plt.subplots(figsize=(7, 6))
ax.plot(df_1['t of A2'], df_1['F(x)(CDF)'], label='Theoretical $F(x)$', color='black', linestyle='-')
ax.plot(df_1['t of A2'], df_1['Sn(x)(PDF)'], label='Empirical $S_n(x)$ (A2)', color='blue', linestyle='--')
ax.plot(df_1['t of A2'], df_1['F(x)(CDF).1'], label='Empirical $S_n(x)$ (A1)', color='red', linestyle='-.')
ax.legend(frameon=False, loc='lower right')

ax.set_xlabel('Time $t$ (months)')
ax.set_ylabel('Cumulative Distribution Function (CDF)')
ax.set_title('Comparison between theoretical and empirical distributions\n of the datasets from Areas A1 and A2.')
ax.set_xlim([0, 120])
ax.set_ylim([0, 1.1])

for spine in ax.spines.values():
    spine.set_linewidth(1.2)       
    spine.set_color('black')       
    spine.set_visible(True)       

ax.tick_params(direction='in', length=4, width=1, colors='black', labelsize=16)
ax.grid(
    visible=True,
    which='major',
    color='gray',
    linestyle='--',
    linewidth=0.6,
    alpha=0.7
)
ax.minorticks_on()  # เปิด tick ย่อย
ax.grid(
    visible=True,
    which='minor',
    color='lightgray',
    linestyle=':',
    linewidth=0.5,
    alpha=0.6
)
plt.savefig('../sim_result/Fig7.eps', format='eps', bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(df_2['n'],df_2['f(t) of A2'], label='A2', color='red', markersize=3,marker = 'o')
ax.plot(df_2['Date TTF (t) Month'],df_2['f(t) of A1'],label='A1',marker = 's', markersize=3, color='blue', linestyle='--')
ax.legend(frameon=False, loc='lower right')

ax.set_xlabel('Time $t$ (months)')
ax.set_ylabel('PDF, f(t)')
ax.set_title('Probability Density Function, f(t)')
ax.set_xlim([0, 120])
ax.set_ylim([0, 0.020])
plt.savefig('../sim_result/Fig3.eps', format='eps', bbox_inches='tight')
plt.tight_layout()
plt.show()
```
